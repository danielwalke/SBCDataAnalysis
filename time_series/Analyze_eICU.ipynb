## Overview of files and file sizes

```python
import os

FILES_FOLDER = "/datasets/eICU/physionet.org/files/eicu-crd/2.0/"
files_list = os.listdir(os.getcwd()+FILES_FOLDER)

size_of_file = [
    (f,os.stat(os.path.join(f"{os.getcwd()}{FILES_FOLDER}", f)).st_size)
    for f in files_list
]

for f,s in size_of_file:
    print("{} : {}MB".format(f, round(s/(1024*1024),3)))
```
