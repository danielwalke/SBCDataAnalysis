```python
lab_item_ids = []
nan_values = []
for i, column in enumerate(pivot_hemato_df_wo_hadm.columns):
    if i < 2: 
        continue
    lab_item_ids.append(column)
    nan_values.append(pivot_hemato_df_wo_hadm[column].isna().sum())
    print(f"{column} - {str(pivot_hemato_df_wo_hadm[column].isna().sum())}")

lab_nan_df = pd.DataFrame(data={"lab_item_ids": lab_item_ids, "nan_values": nan_values})
```

```python
import numpy as np
most_prev_lab_items = np.squeeze(lab_nan_df.sort_values("nan_values").iloc[:4, :].loc[:, ["lab_item_ids"]].values)
```

```python
import dask.dataframe as dd

labs_dd = dd.from_pandas(pivot_hemato_df_wo_hadm.loc[:, ["subject_id","charttime", *most_prev_lab_items]], chunksize = 1000)
mimic_dd = dd.from_pandas(mimc_data, chunksize = 1000)
```

```python
more_labs = dd.merge(mimic_dd, labs_dd, on="subject_id")
```

```python
more_labs['outtime']= dd.to_datetime(more_labs['outtime'])
more_labs['intime']= dd.to_datetime(more_labs['intime'])
more_labs['charttime_y']= dd.to_datetime(more_labs['charttime_y'])
```

```python
more_labs_red = more_labs[more_labs["charttime_y"] >= more_labs["intime"]]
more_labs_red = more_labs[more_labs["charttime_y"] < more_labs["outtime"]]
more_labs_red = more_labs_red.drop_duplicates(subset = ["subject_id", "charttime_x"])
```

```python
more_labs_red_pd = more_labs_red.compute()
```
