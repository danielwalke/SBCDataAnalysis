```python
import pandas as pd

sbc = pd.read_csv("extdata/sbcdata.csv")
```

```python
sbc.shape
```

```python
import numpy as np

noisy_columns = 100 

noise = np.random.rand(noisy_columns, sbc.shape[0]).squeeze()
```

```python
sbc_with_noise = pd.concat((sbc, pd.DataFrame(np.transpose(noise), columns= [f"Noise{i}" for i in range(noisy_columns)])), axis=1)
print([f"Noise{i}" for i in range(noisy_columns)])
```

```python
sbc_with_noise.to_csv(f"extdata/noisy_sbc_{str(noisy_columns)}.csv")
```
