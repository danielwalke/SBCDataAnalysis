```python
import pandas as pd
from dataAnalysis.Constants import FEATURES
import matplotlib.pyplot as plt

xg_boost = pd.read_csv("XGBoost.csv")
decision_tree = pd.read_csv("Decision tree.csv")
logistig_regression = pd.read_csv("Logistic regression.csv")
random_forest = pd.read_csv("Random forest.csv")
rus_boost = pd.read_csv("RUSBoost.csv")
het_gnn = pd.read_csv("Heterogeneous GNN.csv")
hom_gnn = pd.read_csv("Homogeneous GNN.csv")
neural_network = pd.read_csv("Neural Network.csv")
```

```python
def remove_first_column(df):
    # Was index while df creation
    return df.iloc[: , 1:]

xg_boost = remove_first_column(xg_boost)
decision_tree = remove_first_column(decision_tree)
logistig_regression = remove_first_column(logistig_regression)
random_forest = remove_first_column(random_forest)
rus_boost = remove_first_column(rus_boost)
het_gnn = remove_first_column(het_gnn)
hom_gnn = remove_first_column(hom_gnn)
neural_network = remove_first_column(neural_network)
```

```python
def get_feature_importance(df):
    feature_name, feature_importance = df.columns[1:][i].split(" ")
    
    return f"Feature importance: {feature_importance[1:-1]}"
```

```python
import numpy as np
from dataAnalysis.Constants import STEPS
features = decision_tree.columns[1:]
for i, feature in enumerate(features):
    
    plt.plot(xg_boost.iloc[:,0].values, np.ones(xg_boost.iloc[:,0].values.shape[0])*50, "k--")
    plt.plot(xg_boost.iloc[:,0].values, xg_boost.iloc[:, i+1].values, color="#ff0800", marker="x")
    plt.plot(decision_tree.iloc[:,0].values, decision_tree.iloc[:, i+1].values,color="#f27d79", marker="x")
    plt.plot(logistig_regression.iloc[:,0].values, logistig_regression.iloc[:, i+1].values,color='#877afa', marker="o")
    plt.plot(random_forest.iloc[:,0].values, random_forest.iloc[:, i+1].values, color='#d106ce', marker="x")
    plt.plot(rus_boost.iloc[:,0].values, rus_boost.iloc[:, i+1].values, color='#b83062', marker="x")
    plt.plot(het_gnn.iloc[:,0].values, het_gnn.iloc[:, i+1].values, color="#1500d1", marker="o")
    plt.plot(hom_gnn.iloc[:,0].values, hom_gnn.iloc[:, i+1].values, color="#0290bf", marker="o")
    plt.plot(neural_network.iloc[:,0].values, neural_network.iloc[:, i+1].values, color="#161140", marker="o")
    
    feature_name, feature_importance = decision_tree.columns[1:][i].split(" ")
    
    models = ["No statement possible",
        f"XGBoost {get_feature_importance(xg_boost)}",
             f"Decision tree {get_feature_importance(decision_tree)}",
             f"Logistic regression {get_feature_importance(logistig_regression)}",
             f"Random forest {get_feature_importance(random_forest)}",
              f"RUSBoost {get_feature_importance(rus_boost)}",
             f"Heterogeneous GNN {get_feature_importance(het_gnn)}",
              f"Homogeneous GNN {get_feature_importance(hom_gnn)}",
             f"Neural Network {get_feature_importance(neural_network)}",] ## Featuree importance
    plt.xlabel("Feature ratio [%]")
    plt.ylabel("Ratios of diseased CBC [%]")
    plt.ylim(0, 100)
    plt.xlim(0, 100)
    plt.title(feature_name)
    plt.grid()
    plt.legend(models)
    plt.show()
```
