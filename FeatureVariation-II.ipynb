```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import os

for file in os.listdir("feature-variation-II"):
    df = pd.read_csv(f"feature-variation-II/{file}")
    plt.figure(figsize=(20, 5))
    legend = []
    for column in df.columns:
        if column == "x":
            continue
        plt.plot(df["x"], df[column])
        legend.append(column)
    plt.legend(legend)
    plt.title(file)
    plt.ylim([0, 1])
    plt.show()
```
